```python
# Python program to generate word vectors using Word2Vec

# importing all necessary modules
from gensim.models import Word2Vec
import gensim
from nltk.tokenize import sent_tokenize, word_tokenize
import warnings

warnings.filterwarnings(action='ignore')


# Reads ‘alice.txt’ file
sample = open("/content/sample_data/alice.txt")
s = sample.read()

# Replaces escape character with space
f = s.replace("\n", " ")

data = []

# iterate through each sentence in the file
for i in sent_tokenize(f):
	temp = []

	# tokenize the sentence into words
	for j in word_tokenize(i):
		temp.append(j.lower())

	data.append(temp)

# Create CBOW model
model1 = gensim.models.Word2Vec(data, min_count=1,
								vector_size=100, window=5)

# Print results
print("Cosine similarity between 'alice' " +
	"and 'wonderland' - CBOW : ",
	model1.wv.similarity('alice', 'wonderland'))

print("Cosine similarity between 'alice' " +
	"and 'machines' - CBOW : ",
	model1.wv.similarity('alice', 'machines'))

# Create Skip Gram model
model2 = gensim.models.Word2Vec(data, min_count=1, vector_size=100,
								window=5, sg=1)

# Print results
print("Cosine similarity between 'alice' " +
	"and 'wonderland' - Skip Gram : ",
	model2.wv.similarity('alice', 'wonderland'))

print("Cosine similarity between 'alice' " +
	"and 'machines' - Skip Gram : ",
	model2.wv.similarity('alice', 'machines'))
```

```python
# Python program to generate word vectors using Word2Vec

# importing all necessary modules
from gensim.models import Word2Vec
import gensim
from nltk.tokenize import sent_tokenize, word_tokenize
import warnings

warnings.filterwarnings(action='ignore')

# Reads ‘alice.txt’ file
sample = open("/content/sample_data/alice.txt")
s = sample.read()

# Replaces escape character with space
f = s.replace("\n", " ")

data = []

# iterate through each sentence in the file
for i in sent_tokenize(f):
    temp = []

    # tokenize the sentence into words
    for j in word_tokenize(i):
        temp.append(j.lower())

    data.append(temp)

# Create CBOW model
model1 = gensim.models.Word2Vec(data, min_count=1, vector_size=100, window=5)

# Print results for CBOW model
print("Cosine similarity between 'alice' and 'wonderland' - CBOW : ",
      model1.wv.similarity('alice', 'wonderland'))

print("Cosine similarity between 'alice' and 'machines' - CBOW : ",
      model1.wv.similarity('alice', 'machines'))

# Create Skip Gram model
model2 = gensim.models.Word2Vec(data, min_count=1, vector_size=100, window=5, sg=1)

# Print results for Skip Gram model
print("Cosine similarity between 'alice' and 'wonderland' - Skip Gram : ",
      model2.wv.similarity('alice', 'wonderland'))

print("Cosine similarity between 'alice' and 'machines' - Skip Gram : ",
      model2.wv.similarity('alice', 'machines'))

# Function to get the most similar words
def get_similar_words(word, model, topn=10):
    return set([similar_word for similar_word, _ in model.wv.most_similar(word, topn=topn)])

# Function to calculate Jaccard similarity
def jaccard_similarity(set1, set2):
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return intersection / union

# Calculate Jaccard similarity for 'alice' and 'machines' using CBOW model
set1_cb_machines = get_similar_words('alice', model1)
set2_cb_machines = get_similar_words('machines', model1)
jaccard_sim_cb_machines = jaccard_similarity(set1_cb_machines, set2_cb_machines)
print(f"Jaccard similarity between sets of 'alice' and 'machines' - CBOW: {jaccard_sim_cb_machines:.2f}")

# Calculate Jaccard similarity for 'alice' and 'machines' using Skip Gram model
set1_sg_machines = get_similar_words('alice', model2) # Use model2 (Skip Gram)
set2_sg_machines = get_similar_words('machines', model2) # Use model2 (Skip Gram)
jaccard_sim_sg_machines = jaccard_similarity(set1_sg_machines, set2_sg_machines) # Calculate for Skip Gram
print(f"Jaccard similarity between sets of 'alice' and 'machines' - Skip Gram: {jaccard_sim_sg_machines:.2f}") # Print
```
